```python
!pip install autocorrect
from autocorrect import spell

def normalize(phrase):
    return " ".join([spell(word) for word in phrase.split()])
```

```python
matches = pd.DataFrame([])
matches['match'] = terms[0].apply(nlp)
```

```python
phrase_doc = nlp(normalize("tired"))
print(phrase_doc)

def get_similarity(match_doc):
    return match_doc.similarity(phrase_doc)

matches['score'] = matches['match'].apply(get_similarity)
matches.sort_values(by='score', ascending=False)
```
